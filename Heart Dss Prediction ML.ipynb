#### Compare Age distribution against the Target

```python
#We have to create age bins due to the number of multiple ages we get

# Create age bins
age_bins = [20,30,40,50,60,70,80]
age_labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
df['Age Group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Use crosstab to compare age bins with target
crosstab_result = pd.crosstab( df['Age Group'], df['target'])

# Plot the comparison
ax = crosstab_result.plot(kind='bar', figsize=(6, 4))

# Add data labels
for container in ax.containers:
    ax.bar_label(container, label_type='edge')
    
plt.title('Target Distribution Across Age Bins')
plt.xlabel('Age Group')
plt.ylabel('Proportion')
plt.legend(title='Target', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

#### Age vs Heart Rate(Thalach) for Heart Dss

```python
#Create a figure
plt.figure(figsize=(6,4))

#Scatter with positive examples
plt.scatter(df.age[df.target==1],  #Where Target= 1
            df.thalach[df.target==1],
            c="salmon")

plt.scatter(df.age[df.target==0],  #Where Target= 0
            df.thalach[df.target==0],
            c="lightblue")
#Add some helful info
plt.title("Heart dss in function of Age and Max Heart Rate (Thalach)")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.legend(["Disease", "No Disease"]);
```
